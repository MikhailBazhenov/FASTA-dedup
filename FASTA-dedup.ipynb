```python
# This is a function for finding unique sequences in fasta file

def fasta_dedup(input_file, output_file = ''):


    def only_letters(stk):
        out = ''
        for i in range(len(stk)):
            if stk[i].isalpha():
                out = out + stk[i]
        return out


    file_name = input_file
    if output_file == "":
        output_file = file_name.split('.')[0] + '_results.txt'
    f1 = open(file_name, 'r')
    f2 = open(output_file, 'w')
    all_seq = []
    seq = ''
    name = ''
    for line in f1:
        if line[0] == '>':
            if name != '':
                seq = only_letters(seq)
                seq = seq.upper()
                all_seq.append([name, seq])
            name = line.strip().split(' ')[0][1:]
            seq = ''
            begin = False
        else:
            begin = True
        if begin:
            seq = seq + line.strip()
    seq = only_letters(seq)
    seq = seq.upper()
    all_seq.append([name, seq])

    alleles = []
    out_list = []
    n = 65
    found = False
    for i in range(len(all_seq)):
        found = False
        for j in range(len(alleles)):
            if all_seq[i][1] == alleles[j][1]:
                out_list.append([all_seq[i][0], alleles[j][0]])
                found = True
        if not found:
            alleles.append([chr(n), all_seq[i][1]])
            out_list.append([all_seq[i][0], alleles[-1][0]])
            n += 1

    f2.write('The number of alleles = ' + str(len(alleles)) + '\n' + '\n')

    for i in range(len(out_list)):
        out = out_list[i][0] + '\t' + ':' + '\t' + out_list[i][1]
        f2.write(out + '\n')
    f2.write('\n')

    for i in range(len(alleles)):
        out = '>' + alleles[i][0] + '\n' + alleles[i][1]
        f2.write(out + '\n')

    f1.close()
    f2.close()

    return(alleles, out_list)

# This is a function for finding unique sequences in fasta file
```

```python
fasta_dedup('Grf9-6A_from_10+genomes.fasta')
```
